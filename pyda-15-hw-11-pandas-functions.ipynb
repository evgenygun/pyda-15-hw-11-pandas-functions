# Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
ratings = pd.read_csv('ratings.csv', sep=',')
ratings
```

```python
movie_ratings = ratings.groupby('movieId').mean()[['rating']]
movie_ratings
```

```python
def movie_classification(movie):
    if movie <= 2:
        return 'Низкий рейтинг'
    elif (movie > 2) and (movie <= 4):
        return 'Средний рейтинг'
    else:
        return 'Высокий рейтинг'
```

```python
movie_ratings['class'] = movie_ratings['rating'].apply(movie_classification)
movie_ratings
```

# Задание 2
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

Результат классификации запишите в отдельный столбец region.

```python
keywords = pd.read_csv('keywords.csv', sep=',')
keywords
```

```python
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}
```

```python
def region_classification(request):
    
    region = ''
    
    for word in request.split():
        if word in sum(geo_data.values(),[]):
            region = word
    
    if region is '':
        return 'undefined'
    else:
        for reg in geo_data:
            if region in geo_data[reg]:
                return reg
```

```python
keywords['region'] = keywords['keyword'].apply(region_classification)
```

Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов.

```python
movies = pd.read_csv('movies.csv', sep=',')
movies
```

В переменную years записываем список из всех годов с 1950 по 2010.

```python
years = list(range(1950, 2011))
```

Создаем функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

1. для каждой строки пройдите по всем годам списка years
2. если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
3. если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год

*Вариант функции согласно описанию в задании:*

```python
def production_year(movie_title):
    
    production_year = 1990
    
    for year in years:
        if str(year) in movie_title:
            production_year = year
    
    return production_year
```

*Вариант функции с использованием регулярных выражений:*

```python
import re
```

```python
def production_year_regex(movie_title):
    
    default_year = 1990
    
    # если результат поиска в названии фильма паттерна из четырех цифр пустой, 
    # тогда функция возвращает значение default_year
    if not re.findall(r'\d{4}', movie_title):
        return default_year
    
    # если результат поиска в названии фильма паттерна из четырех цифр не пустой, 
    # тогда функция возвращает значение найденного года
    else:
        return int(re.findall(r'\d{4}', movie_title)[0])
```

Записываем год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’

```python
movies['year'] = movies['title'].apply(production_year)
movies
```
